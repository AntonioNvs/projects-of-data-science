# Quarto Dia

```python
dados_resultado = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/imersao-dados-desafio-final/main/Dados/dados_resultados.csv')
dados_resultado.head()
```

```python
dados_resultado['acat_inhibitor'].unique()
```

```python
contagem_moa = dados_resultado.select_dtypes('int64').sum().sort_values(ascending=False)

classes = {}

for key, value in contagem_moa.to_dict().items():
  typeClass = key.split('_')[-1]

  try:
    classes[typeClass] += value
  except:
    classes[typeClass] = value

sorted(classes.items(), key=lambda item: item[1], reverse=True)
```

```python
dados_resultado['n_moa'] = dados_resultado.drop('id', axis=1).sum(axis=1)
```

```python
dados_resultado['ativo_moa'] = (dados_resultado['n_moa'] != 0)
```

```python
dados_combinados = pd.merge(dados, dados_resultado[['id', 'n_moa', 'ativo_moa']], on='id')
dados_combinados.head()
```

```python
dados_combinados.query('tratamento == "com_controle"')
```

```python
dados_combinados['eh_controle'] = (dados_combinados['tratamento'] == 'com_controle')
dados_combinados['eh_24'] = (dados_combinados['tempo'] == 24)
dados_combinados['eh_48'] = (dados_combinados['tempo'] == 48)
dados_combinados['eh_72'] = (dados_combinados['tempo'] == 72)
dados_combinados.head()
```

```python
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
```

```python
composto_principal = dados_combinados['composto'].value_counts().index[:5]
plt.figure(figsize=(10, 8))
sns.boxplot(data = dados_combinados.query('composto in @composto_principal'), y = 'g0', x = 'composto', hue='ativo_moa')
```

# Quinto Dia

```python
dados_resultado.head()
```

```python
resultados_das_classificacoes = {}
random_state = 20202
```

```python
from sklearn.model_selection import train_test_split

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, random_state=random_state)
```

```python
from sklearn.linear_model import LogisticRegression

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
resultados_das_classificacoes['LogisticRegression'] = modelo_rlogistica.score(x_teste, y_teste)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
resultados_das_classificacoes['DummyClassifier'] = accuracy_score(y_teste, previsao_dummy)
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, random_state=random_state)

modelo_arvore = DecisionTreeClassifier()
modelo_arvore.fit(x_treino, y_treino)
resultados_das_classificacoes['DecisionTreeClassifier'] = modelo_arvore.score(x_teste, y_teste)
```

```python
from sklearn.ensemble import RandomForestClassifier


x = dados_combinados.drop(['id', 'n_moa', 'ativo_moa', 'composto'], axis=1)
x = pd.get_dummies(x, columns=['tratamento', 'dose', 'tempo'])
y = dados_combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, random_state=random_state)

modelo_randomforest = RandomForestClassifier()
modelo_randomforest.fit(x_treino, y_treino)
resultados_das_classificacoes['RandomForestClassifier'] = modelo_randomforest.score(x_teste, y_teste)
```
